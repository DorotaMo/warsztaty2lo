```python
def predict_survival(pclass, sex, age, sibsp, parch, fare):
    sex_encoded = le.transform([sex])[0]
    data = [[pclass, sex_encoded, age, sibsp, parch, fare]]
    predictions = model.predict(data)[0]
    return "Przeżyje" if predictions == 1 else "Nie przeżyje"

iface = gr.Interface(
    fn=predict_survival,
    inputs=[
        gr.Dropdown([1,2,3], label="Klasa Pasażera (Pclass)"),
        gr.Dropdown(['male','female'], label="Płeć"),
        gr.Number(label="Wiek"),
        gr.Number(label="Ilość rodzeństwa / małżonków (SibSp)"),
        gr.Number(label="Ilość rodziców / dzieci (Parch)"),
        gr.Number(label="Opłata (Fare)")
    ],
    outputs=gr.Textbox(label="Prognoza"),
    title="Titanic Survival Predictor",
    description="Wprowadź dane pasażera, aby przewidzieć, czy przeżyje katastrofę Titanic."
)

iface.launch()
```
